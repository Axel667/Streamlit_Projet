```python
import requests
import pandas as pd

# Initialiser la liste pour collecter toutes les données
all_data = []

# Paramètres de pagination
page = 0
page_size = 1000  # Nombre de lignes par page (maximum autorisé par l'API)
max_pages = 100  # Limite de pages à récupérer

while page < max_pages:
    # Faire la requête avec le paramètre de pagination
    response = requests.get(
        "https://huggingface.co/api/models",
        params={"limit": page_size, "skip": page * page_size},
        headers={}
    )
    
    # Vérifier que la requête est réussie
    if response.status_code == 200:
        data = response.json()
        
        # Ajouter les données de la page à la liste générale
        all_data.extend(data)
        
        # Vérifier s'il y a encore des données à récupérer
        if len(data) < page_size:  # Dernière page atteinte
            break
        else:
            page += 1  # Passer à la page suivante
            
    else:
        print(f"Erreur lors de la requête : {response.status_code}")
        break

# Extraire les informations et les convertir en DataFrame
metrics_data = [{"_id": metric.get("_id"), "ID": metric.get("id"), "Description": metric.get("description")} for metric in all_data]
df = pd.DataFrame(metrics_data)

# Afficher le DataFrame
print(df)
```

```python
import requests
import time

def fetch_all_models():
    all_models = []
    limit = 100  # Nombre de modèles par page
    skip = 0     # Offset initial

    while True:
        try:
            response = requests.get(
                'https://huggingface.co/api/models',
                params={'limit': limit, 'skip': skip},
                timeout=10  # Délai d'attente de 10 secondes
            )
            response.raise_for_status()
            models = response.json()
            if not models:
                break
            all_models.extend(models)
            skip += limit
            time.sleep(1)  # Pause d'une seconde entre les requêtes
        except requests.exceptions.HTTPError as http_err:
            if response.status_code == 504:
                print(f"Erreur 504 détectée. Nouvelle tentative dans 5 secondes...")
                time.sleep(5)
                continue
            else:
                print(f"Erreur HTTP: {http_err}")
                break
        except requests.exceptions.RequestException as err:
            print(f"Erreur de requête: {err}")
            break

    return all_models

# Exemple d'utilisation
models = fetch_all_models()
print(f"Nombre total de modèles récupérés : {len(models)}")
```

```python
import requests
import pandas as pd
import streamlit as st
import plotly.express as px

# Charger le CSS personnalisé
with open("style.css") as f:
    st.markdown(f"<style>{f.read()}</style>", unsafe_allow_html=True)

# Afficher le logo Hugging Face
st.image("https://huggingface.co/front/assets/huggingface_logo-noborder.svg", width=50)

# Titre principal
st.markdown("<h1 class='title'>Catalogue des Modèles Hugging Face</h1>", unsafe_allow_html=True)
st.markdown("<p class='subtitle'>Ce tableau présente les informations principales de chaque modèle :</p>", unsafe_allow_html=True)

# Couleur personnalisée pour les graphiques
couleur_principale = "#ffcc00"

# Utiliser la mise en cache pour éviter de recharger les données à chaque exécution
@st.cache_data(ttl=86400)  # Cache les données pendant 24 heures
def load_data():
    all_models_data = []
    offset = 0
    limit = 1000  # Nombre maximum de résultats par requête

    while True:
        response = requests.get(
            "https://huggingface.co/api/datasets",
            params={"limit": limit, "full": "True", "config": "True", "offset": offset},
            headers={}
        )

        if response.status_code == 200:
            data = response.json()

            if not data:
                break  # Plus de données à récupérer

            # Extraire les informations principales de chaque modèle
            for model in data:
                model_info = {
                    "ID": model.get("id"),
                    "Auteur": model.get("author"),
                    "Dernière modification": model.get("lastModified"),
                    "Date de création": model.get("createdAt"),
                    "Likes": model.get("likes"),
                    "Téléchargements": model.get("downloads"),
                    "Tags": model.get("tags"),
                    "Description": model.get("description"),
                    "Langues": model.get("cardData", {}).get("language"),
                    "Taille des catégories": model.get("cardData", {}).get("size_categories")
                }
                all_models_data.append(model_info)

            offset += limit  # Passer au prochain ensemble de données
        else:
            print(f"Erreur lors de la requête : {response.status_code}")
            break  # Arrêter la boucle en cas d'erreur

    df = pd.DataFrame(all_models_data)
    return df

# Afficher un message de chargement pendant le chargement des données
with st.spinner('Chargement des données, veuillez patienter...'):
    df = load_data()

# Vérifier le nombre de doublons dans la colonne 'ID'
nombre_doublons_id = df['ID'].duplicated().sum()
print(f"Nombre de doublons dans la colonne 'ID' : {nombre_doublons_id}")

# Convertir la colonne 'Date de création' en format datetime et gérer les valeurs invalides
df['Date de création'] = pd.to_datetime(df['Date de création'], errors='coerce')
df = df.dropna(subset=['Date de création'])

# Grouper par mois et compter les modèles créés chaque mois
df['Mois'] = df['Date de création'].dt.to_period('M')
monthly_counts = df.groupby('Mois').size().reset_index(name='Nombre de modèles')
monthly_counts['Mois'] = monthly_counts['Mois'].dt.to_timestamp()

# Afficher le graphique de tendance des modèles créés par mois avec la couleur principale
st.markdown("<h2 class='subtitle'>Nombre de modèles créés par mois</h2>", unsafe_allow_html=True)
fig_time_series = px.line(
    monthly_counts,
    x='Mois',
    y='Nombre de modèles',
    title="Nombre de modèles créés par mois",
    labels={"Mois": "Date", "Nombre de modèles": "Nombre de modèles créés"},
    color_discrete_sequence=[couleur_principale]
)
fig_time_series.update_traces(line=dict(color=couleur_principale))
st.plotly_chart(fig_time_series)

# Générer le violon plot pour la colonne 'Likes' avec la couleur principale
st.markdown("<h2 class='subtitle'>Distribution des Likes</h2>", unsafe_allow_html=True)
fig_likes = px.violin(df, y="Likes", box=True, points="all", title="Distribution des Likes")
fig_likes.update_traces(marker_color=couleur_principale, line_color=couleur_principale)
st.plotly_chart(fig_likes)
```

```python
import requests
import pandas as pd

# Initialisation de la DataFrame
df = pd.DataFrame()

# Configuration de la pagination
limit = 1000  # Nombre maximum d'éléments par page
offset = 0    # Début de l'index pour chaque requête

while True:
    # Faire la requête à l'API avec pagination
    response = requests.get(
        "https://huggingface.co/api/datasets",
        params={"limit": limit, "offset": offset, "full": "True", "config": "True"},
        headers={}
    )
    
    # Vérifier que la requête est réussie
    if response.status_code == 200:
        data = response.json()  # Récupérer le contenu JSON de la réponse
        
        # Arrêter si aucune donnée n'est reçue (fin des pages)
        if not data:
            break
        
        # Extraire les informations principales de chaque modèle dans une liste de dictionnaires
        models_data = []
        for model in data:
            model_info = {
                "ID": model.get("id"),
                "Auteur": model.get("author"),
                "Dernière modification": model.get("lastModified"),
                "Date de création": model.get("createdAt"),
                "Likes": model.get("likes"),
                "Téléchargements": model.get("downloads"),
                "Tags": model.get("tags"),
                "Description": model.get("description"),
                "Langues": model.get("cardData", {}).get("language"),
                "Taille des catégories": model.get("cardData", {}).get("size_categories")
            }
            models_data.append(model_info)
        
        # Ajouter les données de la page actuelle dans le DataFrame
        df = pd.concat([df, pd.DataFrame(models_data)], ignore_index=True)
        
        # Mettre à jour l'offset pour la prochaine page
        offset += limit
    else:
        print(f"Erreur lors de la requête : {response.status_code}")
        break

# Afficher le DataFrame final
print(df)
```

```python
import threading
import socket
import requests
import json

def data_fetcher(sock):
    limit = 1000
    offset = 0
    while True:
        response = requests.get(
            "https://huggingface.co/api/datasets",
            params={"limit": limit, "offset": offset, "full": "True", "config": "True"},
            headers={}
        )
        if response.status_code == 200:
            data = response.json()
            if not data:
                break
            for model in data:
                model_info = {
                    "ID": model.get("id"),
                    "Auteur": model.get("author"),
                    "Dernière modification": model.get("lastModified"),
                    "Date de création": model.get("createdAt"),
                    "Likes": model.get("likes"),
                    "Téléchargements": model.get("downloads"),
                    "Tags": model.get("tags"),
                    "Description": model.get("description"),
                    "Langues": model.get("cardData", {}).get("language"),
                    "Taille des catégories": model.get("cardData", {}).get("size_categories")
                }
                message = json.dumps(model_info) + '\n'
                sock.sendall(message.encode('utf-8'))
            offset += limit
        else:
            print(f"Erreur lors de la requête : {response.status_code}")
            break
    sock.close()

def start_server():
    serversocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    host = "localhost"
    port = 9999
    serversocket.bind((host, port))
    serversocket.listen(1)
    print(f"Serveur en écoute sur {host}:{port}")
    clientsocket, addr = serversocket.accept()
    print(f"Connexion établie avec {addr}")
    data_fetcher(clientsocket)

# Démarrer le serveur dans un thread séparé
threading.Thread(target=start_server).start()

# Le reste de votre code peut continuer à s'exécuter ici sans être bloqué
print("Le code principal continue son exécution...")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StringType, IntegerType, ArrayType
from pyspark.sql.functions import from_json, col

spark = SparkSession.builder.appName("SocketStream").getOrCreate()

# Définir le schéma des données entrantes
schema = StructType() \
    .add("ID", StringType()) \
    .add("Auteur", StringType()) \
    .add("Dernière modification", StringType()) \
    .add("Date de création", StringType()) \
    .add("Likes", IntegerType()) \
    .add("Téléchargements", IntegerType()) \
    .add("Tags", ArrayType(StringType())) \
    .add("Description", StringType()) \
    .add("Langues", ArrayType(StringType())) \
    .add("Taille des catégories", ArrayType(StringType()))

# Lire les données du socket en streaming
lines = spark \
    .readStream \
    .format("socket") \
    .option("host", "localhost") \
    .option("port", 9999) \
    .load()

# Parser les données JSON
json_df = lines.select(from_json(col("value"), schema).alias("data")).select("data.*")

# Traitement des données (par exemple, afficher dans la console)
query = json_df.writeStream \
    .outputMode("append") \
    .format("console") \
    .start()

query.awaitTermination()
```

```python
import requests
import pandas as pd
import time

# Initialisation des variables
offset = 0
limit_per_request = 50  # Limite de modèles par requête

# Boucle de pagination avec affichage au fur et à mesure
while True:
    response = requests.get(
        "https://huggingface.co/api/models",
        params={"limit": limit_per_request, "offset": offset, "full": "True", "config": "True"},
        headers={}
    )

    if response.status_code == 200:
        data = response.json()
        
        # Si aucune donnée n'est renvoyée, on arrête la boucle
        if not data:
            break
        
        # Récupération des informations et affichage immédiat
        models_data = []
        for model in data:
            model_info = {
                "ID": model.get("id"),
                "Auteur": model.get("author"),
                "Dernière modification": model.get("lastModified"),
                "Date de création": model.get("createdAt"),
                "Likes": model.get("likes"),
                "Téléchargements": model.get("downloads")
            }
            models_data.append(model_info)
        
        # Affichage des données de la requête en cours avec flush pour sortie immédiate
        df = pd.DataFrame(models_data)
        print(df, flush=True)  # flush=True force l'affichage immédiat du print
        
        # Augmente l'offset pour la prochaine requête
        offset += limit_per_request

        # Pause pour éviter la surcharge de requêtes
        time.sleep(1)  # Pause d'une seconde entre chaque requête
    else:
        print(f"Erreur lors de la requête : {response.status_code}", flush=True)
        break
```
